```python
import os
import csv
import shutil

path = r"D:\Data\ShapeNet\models-screenshots\screenshots"
object_name = os.listdir(path)
# labels = ['bathtub', 'bed', 'chair', 'desk', 'dresser', 'monitor', 'night stand', 'sofa', 'table', 'toilet']
labels = ['airplane', 'bathtub', 'bed', 'bench', 'bookshelf', 'bottle', 'bowl', 'car', 'chair', 'cone', 'cup', 'curtain', 'desk', 'door', 'dresser', 'flower pot', 'glass box', 'guitar', 'keyboard', 'lamp', 'laptop', 'mantel', 'monitor', 'night stand', 'person', 'piano', 'plant', 'radio', 'range hood', 'sink', 'sofa', 'stairs', 'stool', 'table', 'tent', 'toilet', 'tv stand', 'vase', 'wardrobe', 'xbox']

save_dir = r"D:\Data\ShapeNet\models-screenshots\ShapeNet40"

with open(r"D:\Data\ShapeNet\metadata.csv", "r") as label_file:
    lines = csv.DictReader(label_file)
    for i, line in enumerate(lines):
        for l in labels:
            if l in line['wnlemmas'] or l in line['name'] or l in line['tags']:
                sel_obj = line['fullId'][len('wss.'):]
                savepath = os.path.join(save_dir, l, f'{i}')
                
                shutil.copytree(os.path.join(path, sel_obj), savepath, dirs_exist_ok=True)
```

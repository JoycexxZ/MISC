```python
import os

logs = os.listdir('.')
logs = [n for n in logs if '.txt' in n]
for log in logs:
    print(log)
    with open(log, "r") as f:
        lines = f.readlines()
        epoch = 0
        num_iter = 0
        d = {key: {"0.25": 0.0, "0.5":0.0} for key in range(3)}
        for line in lines:
            if "INFO: Train: " in line:
                parsed = line[len("[09/11 02:13:26] group-free INFO: Train: "):]
                parsed = parsed.replace("]", "[").split("[")
                num_iter += 1
                if epoch != eval(parsed[1]):
                    epoch = eval(parsed[1])
                    print(''.join([f"epoch {epoch} - "] + [f"layer{i} {d[i]['0.25']/num_iter:.4f} {d[i]['0.5']/num_iter:.4f}\t" for i in range(3)]))
                    d = {key: {"0.25": 0.0, "0.5":0.0} for key in range(3)}
                    num_iter = 0
                    
            if "reflast_train_acc_0.25" in line:
                parsed = line[len("[09/11 00:57:12] group-free INFO: "):]
                parsed = parsed.split(" ")
                d[2]["0.25"] += eval(parsed[1])
                d[2]["0.5"] += eval(parsed[3])
                
            if "1refhead_train_acc_0.25" in line:
                parsed = line[len("[09/11 00:57:12] group-free INFO: "):]
                parsed = parsed.split(" ")
                d[1]["0.25"] += eval(parsed[1])
                d[1]["0.5"] += eval(parsed[3])
            
            if "0refhead_train_acc_0.25" in line:
                parsed = line[len("[09/11 00:57:12] group-free INFO: "):]
                parsed = parsed.split(" ")
                d[0]["0.25"] += eval(parsed[1])
                d[0]["0.5"] += eval(parsed[3])
                
    print('')
```

```python
import numpy as np
with open("sel_num.txt", "r", encoding='utf8') as f:
    lines = f.readlines()
    lines = ''.join(lines).split('tensor')
    lines = [l for l in lines if 'device' in l]
    lines = [l.replace(')', '(').replace(', device','(').split('(') for l in lines]
    num = np.concatenate([eval(l[1]) for l in lines])
    print(num.mean())
```

```python
with open("a.txt", "r") as f:
    a_lines = f.readlines()

with open("b.txt", "r") as f:
    b_lines = f.readlines()
    
a_acc = [eval(l.split(" ")[-1][:-2]) for l in a_lines]
b_acc = [eval(l.split(" ")[-1][:-2]) for l in b_lines]
cn = [l.replace(")", "(").split("(")[1] for l in a_lines]
total = [int(eval(l.replace("correct", "total").split("total: ")[1])) for l in a_lines]
for i in range(len(cn)):
    print("{:15} {:8d} {:8.2f} {:8.2f} {:10.4f}".format(cn[i], total[i], a_acc[i], b_acc[i], a_acc[i]-b_acc[i]))
```

```python
import os
import shutil
root = "/remote-home/share/Datasets/ModelNet40_new2/"
root2 = "/remote-home/share/Datasets/ModelNet40/"

classnames = os.listdir(root)
for classname in classnames:
    full_off = os.listdir(os.path.join(root2, classname, "train"))
    new_off = os.listdir(os.path.join(root, classname, "train"))
    for off in full_off:
        if off not in new_off:
            shutil.copyfile(os.path.join(root2, classname, "train", off), os.path.join(root, classname, "train", off))
    
```

```python
import os
lists = sorted(os.listdir("/remote-home/shicheng/3d/modelnet_render/safe_list/"))
write_lines = []
for l in lists:
    with open(os.path.join("/remote-home/shicheng/3d/modelnet_render/safe_list/", l), "r") as f:
        lines = f.readlines()
    lines = [line.strip().replace('.', '/').split("/")[-4:-1] for line in lines]
    lines = [[line[0].replace('_', ' '), line[1], line[2]] for line in lines]
    lines = ['/'.join(line) + '\n' for line in lines]
    write_lines.extend(lines)
    
with open("modelnet40_safe_train_path.txt", "w") as f:
    f.writelines(write_lines)
        
```
